```python
#IMPORTS LIBRARIES FOR DATA ANALYSIS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load user profile dataset
df_users = spark.read.csv(
    "dbfs:/Volumes/workspace/default/user_profile/1774627765266_user_profile.csv",
    header=True,
    inferSchema=True
)

# Load viewership dataset
df_views = spark.read.csv(
    "dbfs:/Volumes/workspace/default/viewership/1774627769622_viewership.csv",
    header=True,
    inferSchema=True
)
```

```python
# Convert Spark DataFrame to Pandas
pdf_users = df_users.toPandas()
pdf_views = df_views.toPandas()
```

```python
# viewing user_profile data
pdf_users.head()
```

```python
#viewing viewership data
pdf_views.head()
```

```python
# Merge both datasets using user_id as the common key
# how='left', it keeps all viewership data, and attach matching user info
# df.head() shows the first 5 rows of the merged dataset
df = pdf_views.merge(pdf_users, on='UserID', how='left')
df.head()
```

```python
# Clean column names (for future steps (duration 2 To duration_2 & social media handle TO social_media_handle))
df.columns = df.columns.str.lower().str.replace(" ", "_")

df.head()
```

```python
#showing the colomns we have
df.columns
```

```python
#Cleaning the dataset
# Drop unnecessary columns
df = df.drop(columns=['_c4', '_c5'])

# Check result
df.head()
```

```python
# Convert to datetime
df['recorddate2'] = pd.to_datetime(df['recorddate2'])

# Convert to SA time (+2 hours)
df['sa_time'] = df['recorddate2'] + pd.Timedelta(hours=2)

# Extract day name (Monday, Tuesday…)
df['day'] = df['sa_time'].dt.day_name()

# Extract hour (0–23)
df['hour'] = df['sa_time'].dt.hour
```
