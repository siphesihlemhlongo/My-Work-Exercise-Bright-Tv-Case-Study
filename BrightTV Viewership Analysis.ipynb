```python
#IMPORTS LIBRARIES FOR DATA ANALYSIS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load user profile dataset
df_users = spark.read.csv(
    "dbfs:/Volumes/workspace/default/user_profile/1774627765266_user_profile.csv",
    header=True,
    inferSchema=True
)

# Load viewership dataset
df_views = spark.read.csv(
    "dbfs:/Volumes/workspace/default/viewership/1774627769622_viewership.csv",
    header=True,
    inferSchema=True
)
```

```python
# Convert Spark DataFrame to Pandas
pdf_users = df_users.toPandas()
pdf_views = df_views.toPandas()
```

```python
# viewing user_profile data
pdf_users.head()
```

```python
#viewing viewership data
pdf_views.head()
```

```python
# Merge both datasets using user_id as the common key
# how='left', it keeps all viewership data, and attach matching user info
# df.head() shows the first 5 rows of the merged dataset
df = pdf_views.merge(pdf_users, on='UserID', how='left')
df.head()
```

```python
# Clean column names (for future steps (duration 2 To duration_2 & social media handle TO social_media_handle))
df.columns = df.columns.str.lower().str.replace(" ", "_")

df.head()
```

```python
#showing the colomns we have
df.columns
```

```python
#Cleaning the dataset
# Drop unnecessary columns
df = df.drop(columns=['_c4', '_c5'])

# Check result
df.head()
```

```python
# Convert to datetime
df['recorddate2'] = pd.to_datetime(df['recorddate2'])

# Convert to SA time (+2 hours)
df['sa_time'] = df['recorddate2'] + pd.Timedelta(hours=2)

# Extract day name (Monday, Tuesday…)
df['day'] = df['sa_time'].dt.day_name()

# Extract hour (0–23)
df['hour'] = df['sa_time'].dt.hour
```

**using import matplotlib.pyplot as plt**

```python
# Views per Day (Bar Chart)
views_per_day = df['day'].value_counts().sort_index()  # Count sessions per day, Mon-Sun order
plt.figure(figsize=(10,6))
views_per_day.plot(kind='bar', color='skyblue')       # Create bar chart
plt.title("BrightTV: Views per Day")                  # Chart title
plt.xlabel("Day of the Week")                         # x-axis label
plt.ylabel("Number of Sessions")                      # y-axis label
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Views per Hour (Line Chart)
views_per_hour = df['hour'].value_counts().sort_index()  # Count sessions per hour
plt.figure(figsize=(12,6))
views_per_hour.plot(kind='line', marker='o', color='orange')  # Line chart with markers
plt.title("BrightTV: Views per Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Sessions")
plt.xticks(range(0,24))
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Top 10 Channels (Bar Chart)
top_channels = df['channel2'].value_counts().head(10)  # Top 10 content channels
plt.figure(figsize=(12,6))
top_channels.plot(kind='bar', color='green')
plt.title("Top 10 BrightTV Channels")
plt.xlabel("Channel")
plt.ylabel("Number of Views")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Gender Distribution (Pie Chart)
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6,6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue','pink','gray'])
plt.title("Gender Distribution of BrightTV Users")
plt.ylabel("")  # remove y-axis label
plt.show()
```

```python
# Age Distribution (Histogram)
plt.figure(figsize=(10,6))
df['age'].hist(bins=15, color='purple', edgecolor='black')  # Histogram with 15 bins
plt.title("Age Distribution of BrightTV Users")
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Users by Province (Bar Chart)
province_counts = df['province'].value_counts()  # Count users per province
plt.figure(figsize=(12,6))
province_counts.plot(kind='bar', color='orange')
plt.title("BrightTV Users by Province")
plt.xlabel("Province")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
